So far, everything looks great. Since this is a practice exercise, lets add the label red/white to respective data sets and then merge them. We can use that to see how will does the kmeans performs. 

```r
#Adding the lables
df1$label <- 'red'
df2$label <- 'white'

#merging the two sets
wine <- merge(df1, df2, all = TRUE)
```

### Data exploration

```r
library(ggplot2)
```

```r
# A histogram of residual sugar presence by wine color
pl <- ggplot(wine,aes(x=residual.sugar)) + geom_histogram(aes(fill=label),color='black',bins=50, position = position_stack(vjust = 0, reverse = T))
# Optional adding of fill colors
pl + scale_fill_manual(values = c('#ae4554','#faf7ea')) + theme_bw()
```

A few white wines have a significantly larger quantity of residual sugar (also can be attributed to less data present (in comparision to white) in the red-wine data)

```r
# A histogram of citric acid content by wine
p2 <- ggplot(wine, aes(citric.acid)) + geom_histogram(aes(fill = label), color = 'black',bins = 50, position = position_stack(vjust = 0, reverse = T))
p2 + scale_fill_manual(values = c('#ae4554','#faf7ea')) + theme_bw()
```

A large number of white wines have more citirc acid content as compared to red wine.

Alcohol content seems to vary from 8% to ~14% in most of the wines.  

```r
#Alcohol content variation
ggplot(wine, aes(label, alcohol, fill = label)) + geom_boxplot(color = 'black') 
```

Alcohol content seems to vary between ~ 8% - 14%. White wine has a slightly higher alcohol content on avergae as compared to red.

```r
#Scatter plot of residual sugar vs citirc acid
pl <- ggplot(wine,aes(x=citric.acid,y=residual.sugar)) + geom_point(aes(color=label),alpha=0.2)
# Optional adding of fill colors
pl + scale_color_manual(values = c('#ae4554','#faf7ea')) +theme_dark()
```

For similar amounts of citric acid, the residual sugar in most white wines is significantly more as comapred to red wine.

```r
#scatter plot of volatile acidity and residual sugar
pl <- ggplot(wine,aes(x=volatile.acidity,y=residual.sugar)) + geom_point(aes(color=label),alpha=0.2)
# Optional adding of fill colors
pl + scale_color_manual(values = c('#ae4554','#faf7ea')) +theme_dark()
```

Red wine seems to have more volatile acidity as compared to white wine.

### Model building

```r
#leaving out the lable to implement k means.
wine.clus <- wine[, 1:12]
```

```r
head(wine.clus)
```

```r
clus.data <- kmeans(wine.clus, centers = 2, iter.max = 20, nstart = 5)
```

```r
clus.data$centers
```

From our exploratory analysis, we can infer that '2' is red and '1' is white as per the cluster means.

```r
table(wine$label, clus.data$cluster)
```

As the matrix shows, the algorithm identifies the red wine cluster more accurately as compared to white wine.
